## Data Discovery and Visualization

1. How does the number of claims differ across locations?
![Viz1](1.png)

There are four possible locations in the dataset. The location with the most claims is Sao Luis (30 claims), then Recife with the second most (25 claims). Third is Fortaleza (22) and the location with the least amount of claims is Natal (21 claims). 



2. What is the distribution of time to close claims?
![Viz2](2.png)

Looking at the time distribution of all claims we can clearly see that most commonly it takes 0 to 1000 days to close a claim. Now let's delve a little deeper. *One outlier (-57) was removed since it represented a negative value.* 

![Viz2_1](2_1.png)

Looking at the more detailed distribution we can confidentaly state that it usually takes anywhere from 0 to 1200 days to close a claim. Most often it takes 100 to 700 days. Let's keep in mind that this is the data for all four of our locations. Some of the timeframes represent quite extreme values. Both of the far right timeframes belong to Sao Luis. Now that we know how the data spreads among all four of the locations let's take a look how the time varies in each specific one. 



3. How does the average time to close claims differ by location?
![Viz3](3.png)

The above chart helps us confirm that the Sao Luis location takes on average the longest to close their claims - 1026 days. The second longest is the Recife location with 861 days. Fortaleza comes third (776 days). Natal closes their claims within 581 days on average, which is the shortest. Looking at this chart we can't miss its similarity with the first chart - number of claims per location. Sao Luis has the most claims and it takes them the longest to close them, while Natal has the least claims and it takes them the least amount of time to close them. Similarly, Recife takes the second longest to close claims while also having the second largest number of claims, while Fortaleza comes third.


Based on the above data we can make the assertion that the more claims one location has, the longer it takes them to close those claims. A root cause analysis should be performed to better understand this trend. However, in the mean time I might suggest that Vivendo consider distributing claims more evenly across all four locations to improve their claims closing times.
